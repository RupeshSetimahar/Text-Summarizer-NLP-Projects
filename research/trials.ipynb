```python
def get_product(x:int,y:int) -> int:
    return x * y
```

```python
get_product(x=2,y=3)
```

```python
get_product(x=2,y="6")  #still works
```

```python
# to get alert from these type of the mistake we use ensure annotations
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_product(x:int,y:int) -> int:
    return x * y
```

```python
get_product(x=2,y=3)
```

```python
get_product(x=2,y="4")
```
